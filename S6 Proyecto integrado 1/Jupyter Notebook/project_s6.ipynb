#### 2.2.1 Decidimos rellenar los valores NaN de la columna 'year_of_release' con el valor 2999 (fecha inexistente) para poder trabajar con los datos

```python
# Rellenando valores NaN para la columna year of release
#df['year_of_release'].fillna(2999,inplace=True)
df['year_of_release']= df['year_of_release'].fillna(2999)
```

```python
#Verificando que no haya más valores NaN
df[df['year_of_release'].isna()]
```

#### 2.2.2 Decidimos eliminar los dos registros que no cuentan con el campo "name" y pertenecen a la plataforma "GEN"debido a que no nos sirven para nuestro estudio, al igual que representan una cantidad muy pequeña de datos que no generará resultados tendenciosos

```python
#Eliminamos los registros de 'name' que sean NaN
df=df[~df['name'].isna()]
df.info()
```

### 2.2.3 Decidimos rellenar los valores NaN de la columna 'rating' con el valor 'UN' (Undetermined), no sería inteligente eliminarlos debido a que son representativos para el estudio.

```python
# Eliminando valores NaN en la columna 'rating
#df['rating'].fillna('UN',inplace=True)
df['rating'] = df['rating'].fillna('UN')
df[df['rating'].isna()]
```

### 2.2.4 Decidimos rellenar los valores NaN de la columna 'critic_score' con el valor numérico 105, esto debido a que es un valor fuera de rango (0-100) y de esta manera podemos filtrarlo para omitirlos en caso de ser necesario, sin perder la posibilidad de analizar el resto de su información, que puede ser representativa para el estudio.

```python
#Eliminando valores NaN en la columna critic_score
#df['critic_score'].fillna(105,inplace=True)
df['critic_score'] = df['critic_score'].fillna(105)
df[df['critic_score'].isna()]
```

### 2.2.4 Decidimos rellenar los valores NaN de la columna 'user_score' con el valor numérico 11, esto debido a que es un valor fuera de rango (0-10) y de esta manera podemos filtrarlo para omitirlos en caso de ser necesario, sin perder la posibilidad de analizar el resto de su información, que puede ser representativa para el estudio.

```python
#Eliminando valores Nan en la columna 'user_score'
#df['user_score'].fillna(11,inplace=True)
df['user_score']=df['user_score'].fillna(11)
df[df['user_score'].isna()]
```

## 5 Probando hipótesis


```python
#Generando un dataframe excluyendo los valores 'tbd' de la columna 'user_score'
df_hyp= df[df['user_score']!='tbd']
df_hyp.head()
```

### 5.1 Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas.

```python
#Generando objetos serie para las plataformas XONE y PC
#Calculando promedios y varianza para las series
xone= df_hyp[df_hyp['platform']=='XOne']['user_score']
pc= df_hyp[df_hyp['platform']=='PC']['user_score']
xone = xone.astype(float)
pc = pc.astype(float)
print("Promedio de calificaciones de usuario de la plataforma XOne: ",xone.mean())
print("Promedio de calificaciones de usuario de la plataforma PC: ",pc.mean())
print()
print("Varianza de calificaciones de usuario de la plataforma XOne: ",np.var(xone))
print("Varianza de calificaciones de usuario de la plataforma PC: ",np.var(pc))


```

```python
#utilizamos la prueba estadística: Hipótesis sobre la igualdad de las medias de dos poblaciones
#Se utiliza la población completa, sin utilizar el rango de los últimos 9 años
alpha=.05
results= st.ttest_ind(xone,pc, equal_var=False)

pvalue=f'{results.pvalue/2:.15f}'
print('valor p: ',pvalue)

if results.pvalue/2 < alpha:
    print('Rechazamos la hipótesis nula')
    
else:
    print('No rechazamos la hipótesis nula')
```

Podemos observar que la media de las calificaciones de usuarios en XOne es de 7.54 y de la PC es de 7.7 

Planteamiento de hipótesis:
* Hipótesis nula: Las calificaciones promedio de los usuarios para las plataformas PC y XOne son iguales 
* Hipótesis alternativa: Las calificaciones promedio para los usuarios de las plataformas PC y XOne son diferentes 


Utilizando la prueba de hipótesis : "Hipótesis sobre la igualdad de las medias de dos poblaciones" observamos que el valor p es de 0.13 por lo tanto:

* No se rechaza la hipótesis Nula, las calificaciones promedio de los usuarios de PC y XOne **son iguales**
* No se tiene suficiente evidencia para afirmar que las calificaciones promedio de los usuarios de PC y XOne son diferentes

Se establece un valor alpha de 0.05 por convención estadística

### 5.2 Las calificaciones promedio de los usuarios para los géneros de Acción y Deportes son diferentes.

```python
#Creando objetos serie para los géneros Action y Sports
#Calculando promedios y varianza para las series
action= df_hyp[df_hyp['genre']=='Action']['user_score']
sports= df_hyp[df_hyp['genre']=='Sports']['user_score']
action = action.astype(float)
sports = sports.astype(float)
print("Promedio de calificaciones de usuario para el género Action: ",action.mean())
print("Promedio de calificaciones de usuario para el género Sports: ",sports.mean())

print()
print("Varianza de calificaciones de usuario para el género Action: ",np.var(action))
print("Varianza de calificaciones de usuario para el género Action: ",np.var(sports))
```

```python
#utilizamos la prueba estadística: Hipótesis sobre la igualdad de las medias de dos poblaciones
#Se utiliza la población completa, sin utilizar el rango de los últimos 9 años
alpha=.05
results= st.ttest_ind(action,sports, equal_var=False)

pvalue=f'{results.pvalue/2:.15f}'
print('valor p: ',pvalue)

if results.pvalue/2 < alpha:
    print('Rechazamos la hipótesis nula')
    
else:
    print('No rechazamos la hipótesis nula')
```

Podemos observar que la media de las calificaciones de usuarios en los géneros Action es de 8.58 y de Sports es de 8.69

Planteamiento de hipótesis:
* Hipótesis nula: Las calificaciones promedio de los usuarios para los géneros Action y Sports son iguales
* Hipótesis alternativa: Las calificaciones promedio para los usuarios para los géneros Action y Sports son diferentes


Utilizando la prueba de hipótesis : "Hipótesis sobre la igualdad de las medias de dos poblaciones" observamos que el valor p es de 0.049 por lo tanto:

* Se rechaza la hipótesis Nula, las calificaciones promedio de los usuarios para los géneros Action y Sports **son diferentes**
* No se tiene evidencia suficiente para afirmar que las calificaciones promedio para los usuarios de Action y Sports son iguales
* Si bien el valor p= 0.049 (muy cerca del umbral 0.05) no es suficiente para fallar a favor de la hipótesis nula. 

Se establece un valor alpha de 0.05 por convención estadística

### 5.3 ¿Cómo formulamos la hipótesis nula y alternativa? / ¿Qué criterio utilizamos  para probar las hipótesis? y ¿Por qué?


La hipótesis nula se crea desde el planteamiento a probar, desafía el status quo, en cambio la hipótesis alternativa se plantea de manera opuesta a la hipótesis a probar, no desafía el status quo y no representa un cambio significativo.

La clave para crear ambas hipótesis está en el planteamiento, si deseamos comparar diferencia (Mayor que / Menor que) o si deseamos comparar igualdad (Diferente de) entre las dos medias a comparar, esto nos dará una pista de que tipo de prueba utilizar (una cola, dos colas, etc.)

* Para el punto 5.1 "Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas " vamos a comparar igualdad, buscamos saber si las medias de los usuarios de las plataformas son diferentes , por lo tanto utilizamos la prueba de "Hipótesis sobre la igualdad de las medias de dos poblaciones"

* Para el punto 5.2 "Las calificaciones promedio de los usuarios para los géneros de Acción y Deportes son diferentes." vamos a comparar también igualdad, buscamos saber si las medias de los usuarios de los géneros son diferentes , por lo tanto utilizamos la prueba de "Hipótesis sobre la igualdad de las medias de dos poblaciones"

* La prueba de "Hipótesis sobre la igualdad de las medias de dos poblaciones" nos ofrece la posibilidad de saber la probabilidad de que las medias de las poblaciones sean iguales (Hipótesis Nula Ho) o diferentes (Hipótesis alternativa H1)

para ambos puntos, la hipótesis alternativa es el opuesto a la hipótesis nula, por lo tanto, se plantearía de la siguiente manera:
* 5.1 Las calificaciones de los usuarios de las plataformas son diferentes
* 5.2 Las calificaciones de los usuarios de los géneros son diferentes
