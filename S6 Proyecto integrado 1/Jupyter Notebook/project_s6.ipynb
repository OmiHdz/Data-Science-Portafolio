#### 2.2.1 Decidimos rellenar los valores NaN de la columna 'year_of_release' con el valor 2999 (fecha inexistente) para poder trabajar con los datos

### 2.2.3 Decidimos rellenar los valores NaN de la columna 'rating' con el valor 'UN' (Undetermined), no ser√≠a inteligente eliminarlos debido a que son representativos para el estudio.

### 2.2.4 Decidimos rellenar los valores NaN de la columna 'critic_score' con el valor num√©rico 105, esto debido a que es un valor fuera de rango (0-100) y de esta manera podemos filtrarlo para omitirlos en caso de ser necesario, sin perder la posibilidad de analizar el resto de su informaci√≥n, que puede ser representativa para el estudio.

```python
#Eliminando valores NaN en la columna critic_score
#df['critic_score'].fillna(105,inplace=True)
df['critic_score'] = df['critic_score'].fillna(105)
df[df['critic_score'].isna()]
```

### 2.2.4 Decidimos rellenar los valores NaN de la columna 'user_score' con el valor num√©rico 11, esto debido a que es un valor fuera de rango (0-10) y de esta manera podemos filtrarlo para omitirlos en caso de ser necesario, sin perder la posibilidad de analizar el resto de su informaci√≥n, que puede ser representativa para el estudio.

```python
#Eliminando valores Nan en la columna 'user_score'
#df['user_score'].fillna(11,inplace=True)
df['user_score']=df['user_score'].fillna(11)
df[df['user_score'].isna()]
```

## 5 Probando hip√≥tesis


```python
#Generando un dataframe excluyendo los valores 'tbd' de la columna 'user_score'
df_hyp= df[df['user_score']!='tbd']
df_hyp.head()
```

### 5.1 Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas.

```python
#Generando objetos serie para las plataformas XONE y PC
#Calculando promedios y varianza para las series
xone= df_hyp[df_hyp['platform']=='XOne']['user_score']
pc= df_hyp[df_hyp['platform']=='PC']['user_score']
xone = xone.astype(float)
pc = pc.astype(float)
print("Promedio de calificaciones de usuario de la plataforma XOne: ",xone.mean())
print("Promedio de calificaciones de usuario de la plataforma PC: ",pc.mean())
print()
print("Varianza de calificaciones de usuario de la plataforma XOne: ",np.var(xone))
print("Varianza de calificaciones de usuario de la plataforma PC: ",np.var(pc))


```

```python
#utilizamos la prueba estad√≠stica: Hip√≥tesis sobre la igualdad de las medias de dos poblaciones
#Se utiliza la poblaci√≥n completa, sin utilizar el rango de los √∫ltimos 9 a√±os
alpha=.05
results= st.ttest_ind(xone,pc, equal_var=False)

pvalue=f'{results.pvalue/2:.15f}'
print('valor p: ',pvalue)

if results.pvalue/2 < alpha:
    print('Rechazamos la hip√≥tesis nula')
    
else:
    print('No rechazamos la hip√≥tesis nula')
```

Podemos observar que la media de las calificaciones de usuarios en XOne es de 7.54 y de la PC es de 7.7 

Planteamiento de hip√≥tesis:
* Hip√≥tesis nula: Las calificaciones promedio de los usuarios para las plataformas PC y XOne son iguales 
* Hip√≥tesis alternativa: Las calificaciones promedio para los usuarios de las plataformas PC y XOne son diferentes 


Utilizando la prueba de hip√≥tesis : "Hip√≥tesis sobre la igualdad de las medias de dos poblaciones" observamos que el valor p es de 0.13 por lo tanto:

* No se rechaza la hip√≥tesis Nula, las calificaciones promedio de los usuarios de PC y XOne **son iguales**
* No se tiene suficiente evidencia para afirmar que las calificaciones promedio de los usuarios de PC y XOne son diferentes

Se establece un valor alpha de 0.05 por convenci√≥n estad√≠stica

### 5.2 Las calificaciones promedio de los usuarios para los g√©neros de Acci√≥n y Deportes son diferentes.

```python
#Creando objetos serie para los g√©neros Action y Sports
#Calculando promedios y varianza para las series
action= df_hyp[df_hyp['genre']=='Action']['user_score']
sports= df_hyp[df_hyp['genre']=='Sports']['user_score']
action = action.astype(float)
sports = sports.astype(float)
print("Promedio de calificaciones de usuario para el g√©nero Action: ",action.mean())
print("Promedio de calificaciones de usuario para el g√©nero Sports: ",sports.mean())

print()
print("Varianza de calificaciones de usuario para el g√©nero Action: ",np.var(action))
print("Varianza de calificaciones de usuario para el g√©nero Action: ",np.var(sports))
```

```python
#utilizamos la prueba estad√≠stica: Hip√≥tesis sobre la igualdad de las medias de dos poblaciones
#Se utiliza la poblaci√≥n completa, sin utilizar el rango de los √∫ltimos 9 a√±os
alpha=.05
results= st.ttest_ind(action,sports, equal_var=False)

pvalue=f'{results.pvalue/2:.15f}'
print('valor p: ',pvalue)

if results.pvalue/2 < alpha:
    print('Rechazamos la hip√≥tesis nula')
    
else:
    print('No rechazamos la hip√≥tesis nula')
```

Podemos observar que la media de las calificaciones de usuarios en los g√©neros Action es de 8.58 y de Sports es de 8.69

Planteamiento de hip√≥tesis:
* Hip√≥tesis nula: Las calificaciones promedio de los usuarios para los g√©neros Action y Sports son iguales
* Hip√≥tesis alternativa: Las calificaciones promedio para los usuarios para los g√©neros Action y Sports son diferentes


Utilizando la prueba de hip√≥tesis : "Hip√≥tesis sobre la igualdad de las medias de dos poblaciones" observamos que el valor p es de 0.049 por lo tanto:

* Se rechaza la hip√≥tesis Nula, las calificaciones promedio de los usuarios para los g√©neros Action y Sports **son diferentes**
* No se tiene evidencia suficiente para afirmar que las calificaciones promedio para los usuarios de Action y Sports son iguales
* Si bien el valor p= 0.049 (muy cerca del umbral 0.05) no es suficiente para fallar a favor de la hip√≥tesis nula. 

Se establece un valor alpha de 0.05 por convenci√≥n estad√≠stica

### 5.3 ¬øC√≥mo formulamos la hip√≥tesis nula y alternativa? / ¬øQu√© criterio utilizamos  para probar las hip√≥tesis? y ¬øPor qu√©?


La hip√≥tesis nula se crea desde el planteamiento a probar, desaf√≠a el status quo, en cambio la hip√≥tesis alternativa se plantea de manera opuesta a la hip√≥tesis a probar, no desaf√≠a el status quo y no representa un cambio significativo.

La clave para crear ambas hip√≥tesis est√° en el planteamiento, si deseamos comparar diferencia (Mayor que / Menor que) o si deseamos comparar igualdad (Diferente de) entre las dos medias a comparar, esto nos dar√° una pista de que tipo de prueba utilizar (una cola, dos colas, etc.)

* Para el punto 5.1 "Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas " vamos a comparar igualdad, buscamos saber si las medias de los usuarios de las plataformas son diferentes , por lo tanto utilizamos la prueba de "Hip√≥tesis sobre la igualdad de las medias de dos poblaciones"

* Para el punto 5.2 "Las calificaciones promedio de los usuarios para los g√©neros de Acci√≥n y Deportes son diferentes." vamos a comparar tambi√©n igualdad, buscamos saber si las medias de los usuarios de los g√©neros son diferentes , por lo tanto utilizamos la prueba de "Hip√≥tesis sobre la igualdad de las medias de dos poblaciones"

* La prueba de "Hip√≥tesis sobre la igualdad de las medias de dos poblaciones" nos ofrece la posibilidad de saber la probabilidad de que las medias de las poblaciones sean iguales (Hip√≥tesis Nula Ho) o diferentes (Hip√≥tesis alternativa H1)

para ambos puntos, la hip√≥tesis alternativa es el opuesto a la hip√≥tesis nula, por lo tanto, se plantear√≠a de la siguiente manera:
* 5.1 Las calificaciones de los usuarios de las plataformas son diferentes
* 5.2 Las calificaciones de los usuarios de los g√©neros son diferentes
